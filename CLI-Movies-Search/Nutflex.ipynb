```python
def ajout_notes() :
    fichierin = open('film.txt',"r")
    with open('notes.csv', 'a+') as csvfile:
        fichierout = csv.writer(csvfile, delimiter=' ',
                                quotechar='|', quoting=csv.QUOTE_MINIMAL)

        v = (str)(fichierin.readlines())
        w = "',}]["
        for char in w:
            v = v.replace(char, "")

        n = int(input("Quelle note voulez-vous donner au film ?"))
        fichierout.writerow([v, n])
        fichierin.close()
        print(v,"a été noté",n, "!")
    main()
```
