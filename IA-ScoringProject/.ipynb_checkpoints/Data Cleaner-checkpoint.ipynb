# MODELLING

## SMOTE

```python
from imblearn.over_sampling import BorderlineSMOTE
from sklearn.model_selection import train_test_split

# SÃ©paration des features (X) et de la variable cible (y)
X = df.drop(['siren', 'scoring'], axis=1)
y = df['scoring']

# CrÃ©ation de l'instance BorderlineSMOTE
smote = BorderlineSMOTE()

# Application de l'oversampling sur chaque classe de la variable cible
X_resampled, y_resampled = smote.fit_resample(X, y)

# VÃ©rification des nouvelles distributions des classes
print("Nouvelles distributions des classes :", y_resampled.value_counts())

# Diviser les donnÃ©es en ensembles d'entraÃ®nement et de test
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=0)

# Afficher la taille des ensembles d'entraÃ®nement et de test
print("Taille de l'ensemble d'entraÃ®nement :", len(X_train))
print("Taille de l'ensemble de test :", len(X_test))
```

## DECISION TREE WITH CROSS VALIDATION 

```python
import mlflow
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

model = DecisionTreeClassifier()

# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=10)

# Fit the GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Print the best parameters and best score
print("Meilleurs hyperparamÃ¨tres:", best_params)
print("Meilleure accuracy:", best_score)
```

```python
# CrÃ©ation du modÃ¨le d'arbre de dÃ©cision
mlflow.set_experiment("Decision Tree with Cross validation")
mlflow.sklearn.autolog()
with mlflow.start_run(run_name = 'Decision Tree maxdepth10 entropy'):
    
    model = DecisionTreeClassifier(
        criterion='gini', 
        max_depth=10, 
        min_samples_leaf=1, 
        min_samples_split=2, 
        random_state=0)

    # Validation croisÃ©e
    scores = cross_val_score(model, X_train, y_train, cv=10, scoring='accuracy')

    # Affichage des scores de validation croisÃ©e
    print("Scores de validation croisÃ©e :", scores)
    print("Moyenne des scores :", scores.mean())

    # EntraÃ®nement du modÃ¨le sur l'ensemble des donnÃ©es d'entraÃ®nement
    model.fit(X_train, y_train)

    # PrÃ©diction sur les donnÃ©es de test
    y_pred = model.predict(X_test)

    # Rapport de classification sur les donnÃ©es de test
    report = classification_report(y_test, y_pred)
    print("Rapport de classification sur les donnÃ©es de test :\n", report)

    # Obtention des poids des features importances avec les noms des features
    feature_importances = model.feature_importances_
    feature_names = X.columns
    importance_with_names = list(zip(feature_names, feature_importances))

    # Tri des feature importances par ordre dÃ©croissant
    importance_with_names_sorted = sorted(importance_with_names, key=lambda x: x[1], reverse=True)

    # PrÃ©paration des donnÃ©es pour le graphique
    labels, importances = zip(*importance_with_names_sorted[:15])  # SÃ©lection des 15 premiers poids importants
    labels = [label.replace("_", " ").title() for label in labels]  # Modification des noms des features

    # CrÃ©ation du graphique Ã  barres
    plt.figure(figsize=(10, 6))
    plt.barh(range(len(importances)), importances)
    plt.yticks(range(len(importances)), labels)
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.title('Poids des Features (15 premiers, triÃ©s par ordre dÃ©croissant)')

    # Inversion de l'ordre du graphique
    plt.gca().invert_yaxis()

    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

# Define the number of training examples to use for the learning curve
train_sizes = np.linspace(0.1, 1.0, 10)

# Calculate the learning curve scores
train_sizes, train_scores, test_scores = learning_curve(model, X_train, y_train, cv=10, train_sizes=train_sizes, scoring='accuracy')

# Calculate the mean and standard deviation of the train scores and test scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plot the learning curve
plt.figure(figsize=(8, 6))
plt.plot(train_sizes, train_mean, 'o-', color='r', label='Training accuracy')
plt.plot(train_sizes, test_mean, 'o-', color='g', label='Test accuracy')
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color='g')
plt.xlabel('Training Examples')
plt.ylabel('Accuracy')
plt.title('Learning Curve')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
import mlflow
import mlflow.sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

# CrÃ©ation du modÃ¨le Random Forest
mlflow.set_experiment("Random Forest with Cross Validation")
mlflow.sklearn.autolog()
with mlflow.start_run(run_name='Random Forest'):

    modelRF = RandomForestClassifier(
        n_estimators=100,
        criterion='gini',
        max_depth=10,
        min_samples_leaf=1,
        min_samples_split=2,
        random_state=0
    )

    # Validation croisÃ©e
    scores = cross_val_score(modelRF, X_train, y_train, cv=10, scoring='accuracy')

    # Affichage des scores de validation croisÃ©e
    print("Scores de validation croisÃ©e :", scores)
    print("Moyenne des scores :", scores.mean())

    # EntraÃ®nement du modÃ¨le sur l'ensemble des donnÃ©es d'entraÃ®nement
    modelRF.fit(X_train, y_train)

    # PrÃ©diction sur les donnÃ©es de test
    y_pred = modelRF.predict(X_test)

    # Rapport de classification sur les donnÃ©es de test
    report = classification_report(y_test, y_pred)
    print("Rapport de classification sur les donnÃ©es de test :\n", report)

    # Obtention des poids des features importances avec les noms des features
    feature_importances = modelRF.feature_importances_
    feature_names = X.columns
    importance_with_names = list(zip(feature_names, feature_importances))

    # Tri des feature importances par ordre dÃ©croissant
    importance_with_names_sorted = sorted(importance_with_names, key=lambda x: x[1], reverse=True)

    # PrÃ©paration des donnÃ©es pour le graphique
    labels, importances = zip(*importance_with_names_sorted[:15])  # SÃ©lection des 15 premiers poids importants
    labels = [label.replace("_", " ").title() for label in labels]  # Modification des noms des features

    # CrÃ©ation du graphique Ã  barres
    plt.figure(figsize=(10, 6))
    plt.barh(range(len(importances)), importances)
    plt.yticks(range(len(importances)), labels)
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.title('Poids des Features (15 premiers, triÃ©s par ordre dÃ©croissant)')

    # Inversion de l'ordre du graphique
    plt.gca().invert_yaxis()

    plt.show()
```
