```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://www.imdb.com/chart/top"
page = requests.get(url)
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
tag = soup.find('div', class_=re.compile(r'by+'))
tag
```
