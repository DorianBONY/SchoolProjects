```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox"
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, 'lxml')
print(soup.prettify)
```

```python
for child in soup.find_all('table')[1].children:
    for td in child:
        print(td.text)
```

```python
tableau = soup.find_all("span",{'class':'champion-object'})
tableau
```

```python
tableau = soup.findAll('table')[0].findAll('tr')
tableau
```

```python
champion = []
for j in soup.find_all("span", {"class":"champion-object"}):
    print(j)
    champion.append(j.text)
print(champion)
```
